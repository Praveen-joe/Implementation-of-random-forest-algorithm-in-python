# Data visualization and implementation of Random Forest in python

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
np.random.seed(0)
```

```python
iris=load_iris()
df=pd.DataFrame(iris.data, columns=['sepal_length','sepal_width','petal_length','petal_width'])
df.head()
```

```python
df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)
df.head(len(df))
```

# Data Visualization

```python
sns.regplot(x=df['sepal_length'],y=df['sepal_width'],data='df',color='red',label='order1',order=1)
sns.regplot(x=df['sepal_length'],y=df['sepal_width'],data='df',color='blue',label='order2',order=2)
sns.regplot(x=df['sepal_length'],y=df['sepal_width'],data='df',color='yellow',label='order3',order=3)
plt.title('Data Analysis')
plt.show()
sns.regplot(x=df['petal_length'],y=df['petal_width'],data='df',color='red',label='order1',order=1)
sns.regplot(x=df['petal_length'],y=df['petal_width'],data='df',color='blue',label='order2',order=2)
sns.regplot(x=df['petal_length'],y=df['petal_width'],data='df',color='yellow',label='order3',order=3)
plt.title('Data Analysis')
plt.show()
```

```python
sns.lmplot(x='sepal_length',y='sepal_width',data=df,hue='species')
plt.title('species')
plt.show()
sns.lmplot(x='petal_length',y='petal_width',data=df,hue='species')
plt.title('species')
plt.show()
```

# Slicing of dataframe: train(75%) & test(25%)

```python
df['is_train']=np.random.uniform(0,1,len(df))<=.75
df.head(len(df))
```

```python
#slicing the dataframe 
train, test=df[df['is_train']==True], df[df['is_train']==False]
print('no of observations in training data:', len(train))
print('no of observations in testing data:', len(test))
```

```python
features=df.columns[:4]
features
```

```python
y=pd.factorize(train['species'])[0]
y
```

# Creating a random forest classifier

```python
clf= RandomForestClassifier(n_jobs=2,random_state=0)
clf.fit(train[features],y)
```

```python
#Predicting the test['species'] by using the 'y'(0,1,2) values of train['species']  
clf.predict(test[features])
```

```python
#predicition probability
clf.predict_proba(test[features])[10:20]
```

```python
#mapping target_names to the predicted values
preds=iris.target_names[clf.predict(test[features])]
```

```python
pd.crosstab(test['species'],preds,rownames=['Actual Species'], colnames=['Predicted Species'])
```
